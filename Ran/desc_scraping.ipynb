```python
counter = 0
good_count = 0
bad_count = 0
for pet in animal_list:
    response = requests.get(pet['url'], headers=headers, cookies=cookies)
    soup = BeautifulSoup(response.text, 'html.parser')
    try:
        pet_desc = soup.find("div", id="story").parent.next_sibling.next_sibling.text
    except Exception as x:
        counter += 1
        bad_count += 1
        print(f"Good:{good_count}/{counter}. Bad:{bad_count}/{counter} || {x}")
        pet['description'] = None
        time.sleep(1)
    else:
        counter += 1
        good_count +=1
        print(f"Good:{good_count}/{counter}. Bad:{bad_count}/{counter} || Good")
        pet['description'] = pet_desc
        time.sleep(1)
```
